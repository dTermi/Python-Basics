```python
# List functions
# Lists are mutable

list1 = ['elem1', 'elem2', 'elem3', 'elem4']
#       [   0        1        2         3  ] ⬅ Index forward
#       [  -8       -7       -6        -5  ] ➡ Index backward

# Fetch a specific element from the list:
print('\nFetch a specific element from the list:', list1[1])

# Fetch consecutive elements from a list:
print('\nFetch consecutive elements from a list:', list1[1:4])

# Mutable lists: Can change elements of the list anytime
print('\nBefore mutating list:', list1)
list1[2] = 'elem5'
print('After mutating list:', list1)

# Reverting the list back:
list1[2] = 'elem3'
print('\nReverting the list back:', list1)

# Add elements to an existing list:
list1.append('elem5')
print('\nAppended list:', list1)

# Add multile elements to an existing list:
list1.append('elem6')
list1.append('elem7')
list1.append('elem8')
print('\nNew appended list:', list1)

noofelem = int(input('Enter how many elements you want to add:'))
for no in range(noofelem):
    last_elem = list1[len(list1)-1] # Fetch the last element in the list 
    num = last_elem.split('m')[1] # Split the last element on the number
    num = int(num) + 1 # Add one to the split number
    list1.append('elem'+str(num)) # Add that element to the list
    num += 1

print('Appended list using for loop:', list1)

list1.insert(13, 'elem14') # Insert an element at a specified index
print('\nList after inserting an element at specified index:', list1)

# List slcing
print('\nComplete list with slicing', list1[::])
print('\nFirst 5 elements after slicing', list1[:5]) # If list1[:n], gets the first (n-1) elements and displays them
print('\nAll elements after a specific element:', list1[6:]) # If list1[n:], gets the last n to lenght - 1 elements and displays them
print('\nSome elements in the middle of the list:', list1[3:8]) # If list1[n:m], gets the n+1 to m indices elements and displays them
```

```python
import numpy as np
import pandas as pd

# DICTIONARY

# Create a function to display all the elements in the dictionary:
def disp_dict():
    for i,j in enumerate(dict1.keys()): # Enumerate: Uses 2 variables to iterate a list/dictionary. The first variable, 'i', fetches the indices of the elements and the second varable, 'j', fetches the value at that indice
        print('Key', i+1, '=', j, ' & Value = ', dict1[j]) # Here, 'i' starts from 0 and 'j' fetches the value of the first entry in the dict

# Dictionary: Stores data in Key-Value pairs. The value can be string, int, bool, list
dict1 = {'Name':'ABC',
        'Age':'55',
        'Height':155,
        'Weight':70}
print('\nTest Dictionary:', dict1)

# Add a new element in the dictionary:
dict1['Sex'] = 'Male'
print('\nDictionary after adding a new element in it: ', dict1)

# Dictionary Length:
print('\nDictionary Length: ', len(dict1))

# Get a spcific value in the dictionary
spec_value = dict1.get('Name') # Fetches the value whose key we have mentioned in the get method
print('\nGet a spcific value in the dictionary: ', spec_value, '\nData type of the specific value: ', type(spec_value))

# Get a list of all the keys in a dictionary
list_keys = dict1.keys()
print('\nGet a list of all the keys in a dictionary:', list_keys, '\nData type of the fetched list of keys: ', type(list_keys))

# Fetch all the values in the dictionary as per your choosing:
disp_dict()

# Update the existing dictionary to add a few more values
dict1.update({'Favorite color':'Black'}) # Cannot add multiple key-value pairs at once in it. Need to create a list and use a loop to add the elements in it or manually add them one by one
print('\nUpdated dictionary after adding one new element in it:')
disp_dict()

# Create 2 new lists for adding elements in the dictionary
list_for_dict_keys = ['Favorite food', 'Favorite music', 'Bespectacled', 'Hair color']
list_for_dict_values = ['Home cooked meal', 'Slow music', True, 'Blond']

# Adding multiple key-value pairs in the dictionary using a for loop:
if len(list_for_dict_keys) == len(list_for_dict_values):
    for a in range(len(list_for_dict_keys)):
        dict1.update({list_for_dict_keys[a]:list_for_dict_values[a]})

    print('\nAfter adding multiple key-value pairs in the dictionary:')
    disp_dict()

list2 = [99,86,99,87,88,99,111,86,103,87,99,86,94,78,77,99,85,86,99]
dict_mode = {}
mode_value = 0

# Mean
num_mean = np.mean(list2)
print('\nMean of List 2:', num_mean)

# Median
num_median = np.median(list2)
print('\nMedian of List 2:', num_median)

# Mode
# run the for loop n number of times where n is the length of the list in which the elements we need to find the mode are 
for num in range(len(list2)):
    '''
    What is happeneing: We are checking if the element selected by the loop, for instance the first then second then third, exists as one of the dictionary's keys. If they exist, then we 
    are fetching the value associated with that key, incrementing it by 1 and updating the existing dictionary with the new value. If the key does not exist in the dictionary, 
    we are creating a new entry in the dictioanry with this key.
    '''
    if list2[num] in dict_mode.keys(): # Check if the nth element of the list is one of the keys of the dictionary
        new_num = dict_mode[list2[num]] + 1 # Creare a new variable and assgn the nth element of the list to that variable
        dict_mode.update({list2[num]:new_num}) # Update the dictionary with the new number:value pair
    else: # If the nth element of the list is NOT one of the keys of the dictionary
        a = 0 # Declare a new variable 'a'
        a += 1 # Increment the variable by 1
        dict_mode[list2[num]] = a

    '''
    Checking which element is the most frequent one in the list: First, check if the dictionary is empty. If the dictionary is empty, the first value will be by default selected as the most 
    frequent element. If the dictionary is not empty, we'll check if the existing number in the variable 'mode_value' is less than the recently selected frequency number. If it is 
    not less, we'll assign the new higher value to the variable 'mode_value'
    '''
    if len(dict_mode) == 0: # Check if dict is empty
        mode_value = dict_mode[list2[num]]

    elif len(dict_mode) != 0 and mode_value < dict_mode[list2[num]]:
        mode_value = dict_mode[list2[num]]
        
print('\nMode of the List 2 is:', mode_value, 'for the number:', list2[num])
```

# TUPLE

```python
'''
Tuple is an ordered sequence of items same as a list. The only difference is that tuples are immutable. Tuples once created cannot be modified.

Tuples are used to write-protect data and are usually faster than lists as they cannot change dynamically.

It is defined within parentheses () where items are separated by commas.
'''

tuple1 = (4, 'test elem', 4+9j, 5, 'something', 88, +7j, 'anything', 44)

# print elements in the tuple
print('All elements in the tuple:')
for i in range(len(tuple1)):
    print('elem', i+1, ':', tuple1[i])

# Tuple slicing:
print('\nFirst 5 Tuple elements:', tuple1[:5])
print('\nElements 3-8 of the tuple:', tuple1[2:8])
```
